```python
#recursive implementation for quick sort
def quicksort(data, l, r):
    #base case of quick sort where we basically have a 1 element already sorted array
    if len(data) == 1:
        return
    #check to ensure that the left and right indices make sense
    if l < r:
        #pick a pivot and partition the array so that all elements to the left of the pivot are < and to the right of the pivot are
        # greater than
        pivot = partition(data, l, r)
        #recursively call quick sort on n/2 elements
        quicksort(data, l, pivot - 1)
        #recursively sort the other n/2 elements 
        quicksort(data, pivot + 1, r)
    return data
    
def partition(data, l, r):
    #setting up my pointers. grabbing i and setting it == left index. pivot is chosen as final element
    i, pivot = l, data[r]
    for j in range(l, r): #setting up second pointer that iterates from entire array. we loop over the first element too to determine whether its > or < pivot and set up i + j accordingly
        if data[j] < pivot: #if the value at jth index is less than the pivot value
            data[j], data[i] = data[i], data[j] # we swap the element at j index with the element at index i which is at the first element of the part of the array which is > partition
            i += 1 #then increment i by 1 to indicate that the part of array < pivot got one more element. basically shifting the > aprt of array an element over
    data[i], data[r] = data[r], data[i] #do one final swap of the pivot swapping it with the first element of hte > half. putting the pivot in the middle between the two sorted arrrays
    return i #returning the index of the pivot 
```

```python
def insertionSort(arr):
    for i in range(1, len(arr)): #set up loop that goes from second element to end of the array essentially looping over unsorted array as the first element is a sorted arrya of size 1
        val = arr[i] # grab the value at the ith index
        hole = i #set the hole = to i. hole is the index that the value will end up in the sorted portion of the array
        #while the hole is within bounds and the value behind the holes location is greater than the value at the hole
        while hole > 0 and arr[hole - 1] > val:
            arr[hole] = arr[hole - 1] #swap that value forward to where the hole index was
            hole -= 1 #hole moves back
        arr[hole] = val #broke out of loop so finally can put the value in the appropriate position in the sorted array
    return arr
```

```python
def mergesort(arr):
    if len(arr) < 2:
        return
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    mergesort(left)
    mergesort(right)
    i, j, k = 0, 0, 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            arr[k] = left[i]
            i += 1
            k += 1
        elif left[i] > right[j]:
            arr[k] = right[j]
            j += 1
            k += 1
    
    while i < len(left):
        arr[k] = left[i]
        i+=1
        k+=1
    
    while j < len(right):
        arr[k] = right[j]
        k += 1
        j += 1 
```

```python
import random
import time
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

times = {}
for x in range(0, 100):
    unsorted = random.sample(range(10000), 1000)
    start_time = time.time()
    quicksort(unsorted, 0, len(unsorted) - 1)
    end_time = time.time()
    times[x] = (end_time - start_time)
    
%matplotlib inline
plt.rcParams["figure.figsize"] = (20, 6)
plt.title("Quick Sort Run Times")
x1, y1 = times.keys(), times.values()

plt.bar(x1, y1);
```

```python
times = {}
for x in range(0, 100):
    unsorted = random.sample(range(10000), 1000)
    start_time = time.time()
    insertionSort(unsorted)
    end_time = time.time()
    times[x] = (end_time - start_time)
    
%matplotlib inline
plt.rcParams["figure.figsize"] = (20, 6)
plt.title("Insertion Sort Run Times")
x1, y1 = times.keys(), times.values()

plt.bar(x1, y1);
```

```python
times = {}
for x in range(0, 100):
    unsorted = random.sample(range(10000), 1000)
    start_time = time.time()
    mergesort(unsorted)
    end_time = time.time()
    times[x] = (end_time - start_time)
    
%matplotlib inline
plt.rcParams["figure.figsize"] = (20, 6)
plt.title("Merge Sort Run Times")
x1, y1 = times.keys(), times.values()

plt.bar(x1, y1);
```
